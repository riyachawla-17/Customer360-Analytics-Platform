```python
# Load from Files
df_crm = spark.read.option("header", "true").csv("Files/crm_customers.csv")
df_txn = spark.read.option("header", "true").csv("Files/customer_transactions.csv")
df_reviews = spark.read.option("header", "true").csv("Files/customer_reviews.csv")

# Save to bronze layer tables
df_crm.write.format("delta").mode("overwrite").save("Tables/bronze_crm_customers")
df_txn.write.format("delta").mode("overwrite").save("Tables/bronze_customer_transactions")
df_reviews.write.format("delta").mode("overwrite").save("Tables/bronze_customer_reviews")
```

```python
df_crm = spark.read.format("delta").load("Tables/bronze_crm_customers")
df_txn = spark.read.format("delta").load("Tables/bronze_customer_transactions")
df_reviews = spark.read.format("delta").load("Tables/bronze_customer_reviews")
```

```python
df_crm_clean = df_crm.dropna(subset=["CustomerID"])
df_txn_clean = df_txn.dropna(subset=["CustomerID"])
df_reviews_clean = df_reviews.dropna(subset=["CustomerID"])
```

```python
df_joined = df_crm_clean.join(df_txn_clean, on = "CustomerID", how = "left")
```

```python
df_joined = df_joined.join(df_reviews_clean, on = "CustomerID", how = "left")
```

```python
df_txn_clean = df_txn_clean.withColumnRenamed("Date", "TransactionDate")
df_reviews_clean = df_reviews_clean.withColumnRenamed("Date", "ReviewDate")
```

```python
df_joined = df_crm_clean.join(df_txn_clean, on="CustomerID", how="left") \
                        .join(df_reviews_clean, on="CustomerID", how="left")
```

```python
df_joined.write.format("delta").mode("overwrite").save("Tables/silver_customer360_clean")
```
