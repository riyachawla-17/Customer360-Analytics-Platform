```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
```

```python
df = spark.read.format("delta").load("Tables/silver_customer360_clean")
```

```python
def simple_sentiment(text):
    if text is None:
        return "Neutral"
    text = text.lower()
    if any(word in text for word in ["love", "excellent", "great", "awesome","good"]):
        return "Positive"
    elif any(word in text for word in ["not", "poor","bad","worst","disappointed"]):
        return "Negative"
    else: 
        return "Neutral"
```

```python
#register udf
sentiment_udf = udf(simple_sentiment, StringType())
```

```python
df_sentiment = df.withColumn("Sentiment", sentiment_udf(df["ReviewText"]))
```

```python
df_sentiment.write.format("delta").mode("overwrite").save("Tables/gold_Customer360_sentiment")
```
